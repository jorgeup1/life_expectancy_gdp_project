## Import Python Modules

First import the preliminary modules that will be used in this project:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Loading the Data

To look for connections between GDP and life expectancy load the datasets into DataFrames so that they can be visualized.

```python
gdp = pd.read_csv("all_data.csv")
print(gdp.head())
```

#### Renaming the Columns

Due to the fact, the column name `Life expectancy at birth (years)` is long, it's necessary to rename it. The new name will be `life_exp`. This new name is easier to handle when coding.

```python
gdp = gdp.rename({"Life expectancy at birth (years)":"life_exp"}, axis = "columns")
gdp.head()
```

#### Line Charts

It's appropiate to use line charts when having time series. In this case, it was analyzed `Year` vs `GDP` and `Year` vs `Life expectancy at birth (years)`.  From the charts below, it can conclude that China and USA have the best growth in the GDP, and in other countries such as Mexico, Chile and Zimbabwe the development is not significant.
Related to life expectancy, Chile, China, Germany, Mexico, and the USA have increased their life expectancy over the years, except Zimbabwe, which had a decrease until 2005. Since that year, it has increased the life expectancy exponentially, although, life expectancy is relatively short compared with the other countries. 

```python
plt.figure(figsize=(8,6))
sns.lineplot(x="Year", y="GDP", hue="Country", data=gdp)
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x="Year", y="life_exp", hue="Country", data=gdp)
plt.ylabel("Life expectancy at birth (years)")
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
plt.show()
```

#### Statistics

It's necessary to know some important statistics such as GDP and Life Expectancy mean.

```python
gdp_means = gdp.drop("Year", axis=1).groupby("Country").mean().reset_index()
gdp_means
```

#### Bar plots

To better understand the mean values of GDP and Life Expectancy per country, bar plots are good tools. We can see that Germany, Chile, and The USA have the highest mean Life Expectancy. Talking about GPD, The USA has the highest GDP and there is a significant difference from the other countries.

```python
ax = sns.barplot(x="Country", y="life_exp", data=gdp_means, order=gdp_means.sort_values("life_exp").Country)
plt.ylabel("Life expectancy at birth (years)")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.show()
```

```python
ax = sns.barplot(x="Country", y="GDP", data=gdp_means, order=gdp_means.sort_values("GDP").Country)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.show()
```

#### Box Plots

Box plots are alternative ways to visualize the distribution of GDP and Life Expectancy values.

```python
plt.figure(figsize=(11,8))
ax = sns.boxplot(x="Country", y="life_exp", data=gdp)
plt.ylabel("Life expectancy at birth (years)")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.show()
plt.clf()

plt.figure(figsize=(11,8))
ax = sns.boxplot(x="Country", y="GDP", data=gdp)
plt.ylabel("GDP in Trillions of U.S. Dollars")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.show()
plt.clf()
```

#### Scatter plots

In order to see if there is any relationship between GDP and Life Expectancy, the best way to achieve that is to draw a scatter plot.
From the scatter got below, it can conclude that, an exception with Zimbabwe, there is a direct relationship between these two variables.

```python
plt.figure(figsize=(11,8))
sns.scatterplot(x="life_exp", y="GDP", hue="Country", data=gdp)
plt.xlabel("Life expectancy at birth (years)")
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.show()
```

#### Histogram plots 

Life Expectancy has a left-skewed distribution, which means that the mean value is less than the mode and median.
GDP has a right-skewed distribution, which means that the mean value is greater than the mode and median.

```python
plt.figure(figsize=(11,8))
sns.histplot(x="life_exp", data=gdp)
plt.xlabel("Life expectancy at birth (years)")
plt.show()
plt.clf()
plt.figure(figsize=(11,8))
sns.histplot(x="GDP", data=gdp)
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.show()
```
